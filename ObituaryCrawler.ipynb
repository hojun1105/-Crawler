```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By

driver = None;

def Initialize():
#driver,options
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.page_load_strategy = 'eager'
    prefs = {"profile.managed_default_content_settings.images": 2}
    options.add_experimental_option("prefs", prefs)
    global driver
    driver = webdriver.Chrome(options=options)
```

# 조선일보 크롤링


```python
def CrawlPageUrlAndAdd(page_number):
    Josun_full_url = f"https://www.chosun.com/national/obituary-personnel/?page={page_number}"
    driver.get(Josun_full_url)
    wait = WebDriverWait(driver,2)
    url_list = []

    #전체페이지에 속하는 기사 각각의 제일 큰범위

    parent_path = '//*[@id="main"]/div[2]/section/div/div/div/div'
    child_path = wait.until(EC.presence_of_all_elements_located((By.XPATH, f"{parent_path}")))
    n = len(child_path)

    for i in range(1, n+1):
        try:
            span_xpath = f'{parent_path}[{i}]/div/div/div/div[1]/div[2]/div[1]/a/span'
            span = wait.until(EC.presence_of_element_located((By.XPATH, span_xpath)))
            if "부음" in span.text:
                link_xpath = f'{parent_path}[{i}]/div/div/div/div[1]/div[2]/div[1]/a'
                link = driver.find_element(By.XPATH, link_xpath)
                href = link.get_attribute('href')
                # url_list.append(href)

                available_url_list.append(href)
        except TimeoutException:
            # 요소를 찾지 못한 경우, 다음 반복으로 넘어갑니다.
            continue
```

```python
def PageCrawl(pageUrl):
    driver.get(pageUrl)
    wait = WebDriverWait(driver,1)
    parent_xpath = '//*[@id="fusion-app"]/div[1]/div[2]/div/section/article/section/p'
    paragraphs = wait.until(EC.presence_of_all_elements_located((By.XPATH, f"{parent_xpath}")))
    crawled_data = []

    for i,p in enumerate(paragraphs):
        text = p.text
        if(text!=''):
            crawled_data_list.append(text)
```

```python
from tqdm import tqdm

available_url_list = []
crawled_data_list = []
def MainFunc():
    Initialize()
    for i in tqdm(range(0,34)):
        CrawlPageUrlAndAdd(i+1)

    print(f"total url count: {len(available_url_list)}")

    for i in tqdm(available_url_list):
        PageCrawl(i)
```

```python
def contains_word_from_list(string, word_list):
    # 문자열을 공백을 기준으로 단어로 분리
    words = string.split()

    # 리스트의 단어 중 하나라도 문자열의 단어에 포함되어 있는지 확인
    return any(word in words for word in word_list)
```

```python
# 유족을 나타내는 키워드들
family_keywords = ['남편상','빙모상','아내상','형님상','오빠상','자녀상','누나상','언니상','부친상', '장인상', '장모상','시부상','조모상','조부상','모친상','시부상']

# 유족 수를 세는 함수
def count_family_members(obituary_list):
    count = 0
    job = []
    for obituary in obituary_list:
        for keyword in family_keywords:
            if keyword in obituary:
                if "·"in obituary:
                    count += obituary.count("·")+1
                else:
                    count += 1

                # if(obituary.count(" ")>2):
                #     print(f"1-{obituary}")
                #     split_data = obituary.split("·")
                #     for split_data_element in split_data:
                #         if(contains_word_from_list(split_data_element,family_keywords)==False):
                #             if(split_data_element.count(" ")>1):
                #                 job.append(split_data_element.split(" ",1)[1])
                #                 print(split_data_element.split(" ",1)[1])
                #         else:
                #             split_twice = split_data_element.split(" ")
                #             print(f"2-{split_twice}")

                #             if(len(split_twice)!=0):
                #                 print(f"3-{split_twice[1:-1]}")
                #                 job.append(split_twice[1:-1])
                # else:
                #     continue
                parts = obituary.rsplit(' ',1)
                print(parts[0])
                job.append(parts[0])
    return (count,job)
```

```python
def ManageColumn(row):
    #이름처리
    new_row = []
    row[0] = row[0].replace("▲","").replace("별세","")
    if(" " in row[0]):
        parts = row[0].split(" ",1)
        new_row.append(parts[0])
        new_row.append(parts[1])
    else:
        new_row.append(row[0])
        new_row.append("Nono")
    #성별
    try:
        if("부친상" in row[1] or "남편상" in row[1] or "조부상" in row[1]):
            new_row.insert(1,"남")
        else:
            new_row.insert(1,"여")
    #유족
        tuple_data = count_family_members(row)
        new_row.append(tuple_data[0])
    #유족수
        new_row.append(tuple_data[1])

    #장례식장
        new_row.append(row[-3].split(" ")[-1])
    except Exception:
        print(f"잘못된 row : {row[0]}")

    return new_row
```

```python
import re

split_data_list = []

def Label(data_list):
    for i in data_list:
        splitString = re.split('[,=]+',i)
        splitString= splitString
        new_row = ManageColumn(splitString)
        split_data_list.append(new_row)
```
